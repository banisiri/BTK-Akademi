```python
class SuperKahraman():
  def __init__(self,isimInput, yasInput, meslekInput):
    print("init cagirildi")
    self.isim=isimInput
    self.yas=yasInput
    self.meslek=meslekInput
  def ornekMethod(self):
    print(f"ben bir super kahramanim ve meslegim:",{self.meslek})
```

```python
superman=SuperKahraman("Clark Kent",30,"gazeteci")
```

```python
superman.isim
```

```python
superman.meslek, superman.yas
```

```python
superman.ornekMethod()
```

```python
class Kopek():

  def __init__(self,yas=5):
    self.yas=yas
  def insanyasi(self):
    return self.yas * 7
```

```python
benimKopek=Kopek()
```

```python
benimKopek.yas
```

```python
benimKopek.insanyasi()
```
