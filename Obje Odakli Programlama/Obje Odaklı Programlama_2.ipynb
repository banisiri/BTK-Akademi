inheritance

```python
class Hayvan():
  def __init__(self):
    print("hayvan sinifi cagirildi")
  def method1(self):
    print("hayvan sinifi method1 caigirildi")
  def method2(self):
    print("hayvan sinifi methodu2 caigirldi")
```

```python
benimHayvanim=Hayvan()
```

```python
benimHayvanim.method1()
```

```python
class Kedi(Hayvan):
  def __init__(self):
    Hayvan.__init__(self)
    print("kedi sinifi init cagirildi")
  def miyavla(self):
    print("miyav")
  #override
  def method1 (self):
    print("kedi sinifindaki method cagirildi")
```

```python
benimKedim=Kedi()
```

```python
benimKedim.method2()
```

```python
benimKedim.miyavla()
```

```python
benimKedim.method1()
```

polymorphism

```python
class Elma():
  def __init__(self,isim):
    self.isim=isim
  def bilgiver(self):
    return self.isim + "100 kaloridir"
```

```python
class Muz():
  def __init__(self,isim):
    self.isim=isim
  def bilgiver(self):
    return self.isim + "150 Kaloridir"
```

```python
elma=Elma("elma")
```

```python
elma.bilgiver()
```

```python
muz=Muz("muz")
```

```python
muz.bilgiver()
```

```python
meyveListesi=[elma,muz]
```

```python
for meyve in meyveListesi:
  print(meyve.bilgiver())
```

```python
def bilgiAl(meyve):
  print(meyve.bilgiver())
```

```python
bilgiAl(muz)
```

```python
bilgiAl(elma)
```
