```python
class Meyve():
  def __init__(self,isim,kalori):
    self.isim=isim
    self.kalori=kalori
  def __str__(self):
    return f"{self.isim} su kadar kaloriye sahiptir:{self.kalori}"
  def __len__(self):
    return self.kalori
```

```python
muz=Meyve("muz",150)
```

```python
muz.kalori
```

```python
print(muz)
```

```python
len(muz)
```
