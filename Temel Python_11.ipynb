```python
def ornekfonksiyon(**kwargs):
  return kwargs
```

```python
def isimkontrol(**kwargs):
  if "sabri" in kwargs:
    print("Sabri var")
  else:
    print("Sabri nerde")
```

```python
isimkontrol(ahmet=10,mehmet=5,sabri=1)
```

```python
isimkontrol(zeynep=3,Sabri=4)
```

```python
kontrolfoksiyonu("b")
```

```python
stringlistesi=["sabri","basri","omer","utku","destan","nazli"]
```

```python
list(map(kontrolfoksiyonu,stringlistesi))
```

```python
map(kontrolfoksiyonu,stringlistesi)
```

```python
sonuclistesi=list(map(kontrolfoksiyonu,stringlistesi))
```

```python
sonuclistesi.count(True)
```
