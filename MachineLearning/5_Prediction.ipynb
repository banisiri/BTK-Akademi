```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
veri=pd.read_csv(r"C:\Users\90507\Desktop\yazılım\Dataset\maaslar.csv")
```

x-->egitim,  y-->maas

```python
x=veri.iloc[:,1:2]
y=veri.iloc[:,2:3]

from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x,y)
```

```python
plt.scatter(x,y,color="b")
plt.plot(x, lr.predict(x),color="r")
```

polynomial regrassion

```python
from sklearn.preprocessing import PolynomialFeatures
ply=PolynomialFeatures(degree=2)
xPly=ply.fit_transform(x)
lr2=LinearRegression()
lr2=lr2.fit(xPly,y)

plt.scatter(x,y,color="b")
plt.plot(x,lr2.predict(ply.fit_transform(x)),color="r")
```

```python
from sklearn.preprocessing import PolynomialFeatures
ply=PolynomialFeatures(degree=8)
xPly=ply.fit_transform(x)
lr2=LinearRegression()
lr2=lr2.fit(xPly,y)

plt.scatter(x,y,color="b")
plt.plot(x,lr2.predict(ply.fit_transform(x)),color="r")
```
