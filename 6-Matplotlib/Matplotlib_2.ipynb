```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
numpyDizisi1=np.linspace(0,10,20)
```

```python
numpyDizisi1
```

```python
numpyDizisi2=numpyDizisi1**2
```

```python
numpyDizisi2
```

```python
(benimFigur,benimEksen)=plt.subplots(dpi=100)
benimEksen.plot(numpyDizisi1,numpyDizisi2,color="#00594F",alpha=0.8)
benimEksen.plot(numpyDizisi2,numpyDizisi1,color="#CEDC00")
```

```python
(yeniFigur,yeniEksen)=plt.subplots(dpi=100)
yeniEksen.plot(numpyDizisi1, numpyDizisi1+2, color="#00594F", linewidth=5, linestyle="-.")
yeniEksen.plot(numpyDizisi1, numpyDizisi1+4, color="yellow", linewidth=3, linestyle=":")
yeniEksen.plot(numpyDizisi1, numpyDizisi1+6, color="green", linestyle="--")

yeniEksen.plot(numpyDizisi1, numpyDizisi1+8, color="#000000", linestyle="--", marker="+", markerfacecolor="red")
```
