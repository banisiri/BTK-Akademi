```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#%matplotlib inline
```

```python
yasListesi=[10,20,30,30,40,50,60,70,75,75]
kiloListesi=[20,60,80,55,67,54,64,76,98,40]
```

```python
numpyYaslistesi=np.array(yasListesi)
numpyKilolistesi=np.array(kiloListesi)
```

```python
# g green b bllue r red
plt.plot(numpyYaslistesi,numpyKilolistesi,"r")
plt.xlabel("yas")
plt.ylabel("kilo")
plt.title("kilo yas grafigi")
```

```python
numpyDizisi1=np.linspace(0,10,20)
numpyDizisi1
```

```python
numpyDizisi2=numpyDizisi1**3
numpyDizisi2
```

```python
plt.plot(numpyDizisi1,numpyDizisi2,"r*-")
```

```python
plt.subplot(1,2,1)
plt.plot(numpyDizisi1,numpyDizisi2,"r*-")
plt.subplot(1,2,2)
plt.plot(numpyDizisi2,numpyDizisi1,"g--")
```

```python
yeniFigurum=plt.figure()
figureAxes=yeniFigurum.add_axes([0.2,0.2,0.6,0.6])
figureAxes.plot(numpyDizisi1,numpyDizisi2,"g")
figureAxes.set_xlabel("X Ekseni verisi")
figureAxes.set_ylabel("Y Ekseni verisi")
figureAxes.set_title("Grafik basligi")
```

```python
figure2=plt.figure()

eksen1=figure2.add_axes([0.1,0.1,0.9,0.9])
eksen2=figure2.add_axes([0.3,0.5,0.3,0.3])

eksen1.plot(numpyDizisi1,numpyDizisi2,"g")
eksen1.set_xlabel("X ekseni")
eksen1.set_ylabel("Y ekseni")
eksen1.set_title("Ana baslik")

eksen2.plot(numpyDizisi1,numpyDizisi2,"b")
eksen2.set_xlabel("X ekseni")
eksen2.set_ylabel("Y ekseni")
eksen2.set_title("Kucuk grafik baslik")
```

```python
(benimfigurum,benimeksenim)=plt.subplots(nrows=1,ncols=2)
```

```python
for eksen in benimeksenim:
    eksen.plot(numpyDizisi1,numpyDizisi2,"g")
plt.tight_layout()
```

```python
benimfigurum
```

```python
yeniFigur=plt.figure(figsize=(5,5),dpi=100)

yeniEksen=yeniFigur.add_axes([0.1,0.1,0.9,0.9])
yeniEksen.plot(numpyDizisi1,numpyDizisi1**2,label="numpyDizisi1**2")
yeniEksen.plot(numpyDizisi1,numpyDizisi1**2.5,label="numpyDizisi1**2.5")
yeniEksen.legend(loc=0)
```
