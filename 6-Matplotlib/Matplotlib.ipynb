```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#%matplotlib inline
```

```python
yasListesi=[10,20,30,30,40,50,60,70,75,75]
kiloListesi=[20,60,80,55,67,54,64,76,98,40]
```

```python
numpyYaslistesi=np.array(yasListesi)
numpyKilolistesi=np.array(kiloListesi)
```

```python
# g green b bllue r red
plt.plot(numpyYaslistesi,numpyKilolistesi,"r")
plt.xlabel("yas")
plt.ylabel("kilo")
plt.title("kilo yas grafigi")
```
